```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib as plt
import seaborn as sns
data = pd.read_csv("apple_products.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
print(data.describe())
```

```python
highest_rated = data.sort_values(by=["Star Rating"], 
                                 ascending=False)
highest_rated = highest_rated.head(10)
print(highest_rated['Product Name'])
```

```python

import matplotlib.pyplot as plt
```

```python

iphones = highest_rated["Product Name"].value_counts()

# The index (labels) are the iPhone models and the values are the count of each model
label = iphones.index
counts = iphones.values

# Create the bar chart using matplotlib
plt.figure(figsize=(10,6))  # Optional: Adjust the figure size for better visibility
plt.bar(label, counts, color='skyblue')

# Set the title and labels
plt.title("Number of Ratings of Highest Rated iPhones")
plt.xlabel("iPhone Model")
plt.ylabel("Number of Ratings")

# Rotate the labels if there are too many models to fit nicely
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()  # Ensures the labels are not cut off
plt.show()
```

```python
label = iphones.index
counts = highest_rated.groupby("Product Name")["Number Of Reviews"].sum().loc[label] 

plt.figure(figsize=(10,6)) 
plt.bar(label, counts, color='skyblue')

plt.title("Number of Reviews of Highest Rated iPhones")
plt.xlabel("iPhone Model")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=45, ha='right')

plt.tight_layout() 
plt.show()
```

```python

plt.figure(figsize=(10,6))
plt.scatter(data["Number Of Ratings"], data["Sale Price"], 
            s=data["Discount Percentage"]*10, alpha=0.5, color='blue')

# Add title and labels
plt.title("Relationship between Sale Price and Number of Ratings of iPhones")
plt.xlabel("Number Of Ratings")
plt.ylabel("Sale Price")

plt.show()
```
